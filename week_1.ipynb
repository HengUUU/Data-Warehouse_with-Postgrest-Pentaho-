```python
df = pd.read_csv("data_1.1.csv")
df = df.to_numpy()
```

```python
x_tr = df[:,0]
y_tr = df[:,-1]
# x_tr = torch.tensor(x_tr,dtype=torch.float32,requires_grad=False)
# y_tr = torch.tensor(y_tr,dtype=torch.float32,requires_grad=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Normalize x for better performance in training
x_mean = np.mean(x_tr)
x_std = np.std(x_tr)
x_normalized = (x_tr - x_mean) / x_std

def cost_func(hyp, y):
    """Calculate mean squared error."""
    return ((hyp - y) ** 2).mean()

def hypo_func(a, x, b):
    """Hypothesis function for linear regression."""
    return a * x + b

def grad_a(hyp, y, x):
    """Gradient of cost with respect to a."""
    return 2 * np.mean((hyp - y) * x)

def grad_b(hyp, y):
    """Gradient of cost with respect to b."""
    return 2 * np.mean(hyp - y)

# Initialize parameters
a, b = 0, 0  # Slope and intercept
lr = 0.1   # Learning rate
num_iterations = 1000  # Number of iterations
cost_values = []

# Training loop
for i in range(num_iterations):
    
    y_pred = hypo_func(a, x_normalized, b)  # Calculate predictions
    cost_val = cost_func(y_pred, y_tr)  # Calculate cost

    # Update parameters
    a -= lr * grad_a(y_pred, y_tr, x_normalized)
    b -= lr * grad_b(y_pred, y_tr)
    # cost value
    cost_values.append(cost_val)
    # Print debugging information every 100 iterations
    if i % 100 == 0:
        print(f"Iteration {i}: Cost = {cost_val}, a = {a}, b = {b}")

# Final parameters and cost
print(f"Final parameters: a = {a}, b = {b}")
print(f"Final cost: {cost_func(hypo_func(a, x_normalized, b), y_tr)}")

# Plotting the results
plt.scatter(x_tr, y_tr, color='blue', label='Data Points')
plt.plot(x_tr, hypo_func(a, x_normalized, b), color='red', label='Fitted Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python
plt.plot([i for i in range(len(cost_values))],cost_values)
plt.show()
```

```python
def hypothesis(a,x,b,c):
    return np.sqrt(x)*a + b*x + c

def gradient_a(y,a,x,b,c):
    h = hypothesis(a,x,b,c)
    grada = (h - y)*np.sqrt(x)
    return 2*np.mean(grada)

def gradient_b(y,a,x,b,c):
    h = hypothesis(a,x,b,c)
    gradb = (h-y)*x
    return 2*np.mean(gradb)

def gradient_c(y,a,x,b,c):
    h = hypothesis(a,x,b,c)
    gradc = (h-y)
    return 2*np.mean(gradc)

def cost_function(y,a,x,b,c):
    h = hypothesis(a,x,b,c)
    cost = (h - y)**2
    return cost.mean()

a,b,c,lr = 1,1,1,0.1
```
